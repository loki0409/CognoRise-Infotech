IMPORTING THE NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

READ THE CSV FILE

```python
titanic = pd.read_csv("/content/tested.csv")
titanic
```

TO CHECK NULL VALUES

```python
titanic.isnull().sum()
```

**DATA CLEANING**

```python
tit = titanic.drop(labels=["Cabin","Name","Ticket"], axis=1)
tit
```

```python
tit.isnull().sum()
```

```python
#Fill the null values of 'Age' column with the mean value
tit["Age"].fillna(tit["Age"].mean(),inplace = True)
```

```python
tit.isnull().sum()
tit
```

```python
tit.isnull().sum()
```

```python
tit.isnull().sum()
```

```python
#Replacing the male and female genders as 0 and 1
tit.replace(['male','female'],[0,1],inplace = True)
tit
```

```python
#Replacing the Embarked class values with 0,1,2
tit.replace(["Q","S","C"],[0,1,2],inplace = True)
tit
```

```python
#Fill the null values of 'fare' column with the mean value
tit["Fare"].fillna(tit["Fare"].mean(),inplace = True)
```

```python
tit.isnull().sum()
```

```python
tit.info()
```

SEPARTING DATA FOR TRAINING AND TESTING

```python
x = tit.iloc[:,1:]

y = tit['Survived']
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

LOGISTIC REGRESSION

```python
model = LogisticRegression()
model.fit(x_train,y_train)


```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
Confusion_matrix_prediction_train = confusion_matrix(y_train,x_train_prediction)
Confusion_matrix_prediction_train
```

```python
Confusion_matrix_prediction = confusion_matrix(y_test,x_test_prediction)
Confusion_matrix_prediction
```

```python
print("Logistic Regression: ")
print("Accuracy Score: ", accuracy_score(x_train_prediction,y_train))
print("Classification Report for training set: ")
print(classification_report(y_train, x_train_prediction))
```

```python
print("Logistic Regression: ")
print("Accuracy Score: ", accuracy_score(x_test_prediction,y_test))
print("Classification Report: ")
print(classification_report(y_test, x_test_prediction))
```

```python
#DECISON TREE CLASSIFICATION
from sklearn.tree import DecisionTreeClassifier
Iris = DecisionTreeClassifier()
Iris_train = Iris.fit(x_train,y_train)
Iris_prediction = Iris.predict(x_test)
```

```python
print("The accuracy of the decision tree classifier is :",{accuracy_score(Iris_prediction,y_test)})
```
